```python
def compare_total_values (baseline_df, df_compare_list):
    #Calculate totals for baseline situation
    indoor_unmet_demand = baseline_df['Unmet_Demand_Indoor_Total[AF]'].sum()
    outdoor_unmet_demand = baseline_df['Unmet_Demand_Outdoor_Total[AF]'].sum()
    ag_unmet_demand = baseline_df['Unmet_Demand_Ag_Total[AF]'].sum()
    groundwater_storage = baseline_df[baseline_df['Date'] == '2070-12-01']['Total_gw_storage'].values[0]
    electric_demand = baseline_df['Total_GwH'].sum()
    hydropower_generation = baseline_df['Total_hydro_generation'].sum()

    increase_indoor_unmet_demand_list = []
    increase_outdoor_unmet_demand_list = []
    increase_ag_unmet_demand_list = []
    increase_groundwater_storage_list = []
    increase_electric_demand_list = []
    increase_hydropower_generation_list = []

    for df in df_compare_list: 
    #calculate totals for comparison situation
        indoor_unmet_demand_compare = df['Unmet_Demand_Indoor_Total[AF]'].sum()
        outdoor_unmet_demand_compare = df['Unmet_Demand_Outdoor_Total[AF]'].sum()
        ag_unmet_demand_compare = df['Unmet_Demand_Ag_Total[AF]'].sum()
        groundwater_storage_compare = df[df['Date'] == '2070-12-01']['Total_gw_storage'].values[0]
        electric_demand_compare = df['Total_GwH'].sum()
        hydropower_generation_compare = df['Total_hydro_generation'].sum()
        
        #calculate outputs
        increase_indoor_unmet_demand = (indoor_unmet_demand_compare - indoor_unmet_demand)
        increase_outdoor_unmet_demand = outdoor_unmet_demand_compare - outdoor_unmet_demand
        increase_ag_unmet_demand = ag_unmet_demand_compare - ag_unmet_demand
        increase_groundwater_storage = groundwater_storage_compare - groundwater_storage
        increase_electric_demand = electric_demand_compare - electric_demand
        increase_hydropower_generation = hydropower_generation_compare - hydropower_generation
    
        #add outputs to lists
        increase_indoor_unmet_demand_list.append(increase_indoor_unmet_demand)
        increase_outdoor_unmet_demand_list.append(increase_outdoor_unmet_demand)
        increase_ag_unmet_demand_list.append(increase_ag_unmet_demand)
        increase_groundwater_storage_list.append(increase_groundwater_storage)
        increase_electric_demand_list.append(increase_electric_demand)
        increase_hydropower_generation_list.append(increase_hydropower_generation)
    
    
    # #print all values
    # print('reduction in supply deliveries indoor', indoor_supply_delivered - indoor_supply_delivered_adapt)
    # print('reduction in supply deliveries outdoor', outdoor_supply_delivered - outdoor_supply_delivered_adapt)
    # print('reduction in supply deliveries ag', ag_supply_delivered - ag_supply_delivered_adapt)

    # print('total coverage indoor percent:', (indoor_supply_delivered / (indoor_supply_delivered + indoor_unmet_demand)))
    # print('total coverage outdoor percent:', (outdoor_supply_delivered / (outdoor_supply_delivered + outdoor_unmet_demand)))
    # print('total coverage ag percent:', (ag_supply_delivered / (ag_supply_delivered + ag_unmet_demand)))

    # print('total coverage indoor adapt percent:', (indoor_supply_delivered_adapt / (indoor_supply_delivered_adapt + indoor_unmet_demand_adapt)))
    # print('total coverage outdoor adapt percent:', (outdoor_supply_delivered_adapt / (outdoor_supply_delivered_adapt + outdoor_unmet_demand_adapt)))
    # print('total coverage ag adapt percent:', (ag_supply_delivered_adapt / (ag_supply_delivered_adapt + ag_unmet_demand_adapt)))

    return  increase_indoor_unmet_demand_list, increase_outdoor_unmet_demand_list, increase_ag_unmet_demand_list, increase_groundwater_storage_list, increase_electric_demand_list, increase_hydropower_generation_list
```

```python
def calculate_indoor_coverage (df):
    indoor_unmet_demand = df['Unmet_Demand_Indoor_Total[AF]'].sum()
    indoor_supply_delivered = df['Supply_Del_Indoor_Total[AF]'].sum()
    indoor_coverage = (indoor_supply_delivered / (indoor_unmet_demand + indoor_supply_delivered))*100
    return indoor_coverage
    
def calculate_outdoor_coverage (df):
    outdoor_unmet_demand = df['Unmet_Demand_Outdoor_Total[AF]'].sum()
    outdoor_supply_delivered = df['Supply_Del_Outdoor_Total[AF]'].sum()
    outdoor_coverage = (outdoor_supply_delivered / (outdoor_unmet_demand + outdoor_supply_delivered))*100
    return outdoor_coverage

def calculate_ag_coverage (df):
    ag_unmet_demand = df['Unmet_Demand_Ag_Total[AF]'].sum()
    ag_supply_delivered = df['Supply_Del_Ag_Total[AF]'].sum()
    ag_coverage = (ag_supply_delivered / (ag_unmet_demand + ag_supply_delivered))*100
    return ag_coverage
```

```python
def graph_number_line (df_list, metric_name):
    # Create a simple number line

    #If coverage is the metric, put these values on a 0 to 100 number line
    if metric_name in ['Urban Indoor Coverage (Percentage)', 'Urban Outdoor Coverage (Percentage)']:
        x = range(85, 101)  # Numbers from 80 to 100
    elif metric_name == 'Agricultural Coverage (Percentage)':
        x = range(50, 81)
    elif metric_name == 'Groundwater Storage in MAF':
        x = range(1100, 1500)
    elif metric_name == 'Electricity Use in TWh':
        x = range(4000, 6000) 
    else:
        x = range(7500, 9000)
    
    y_values = [0, 20, 40, 60, 80, 100]  # List of y-coordinates for the lines
    
    # Plot the number lines
    for y in y_values:
        plt.plot(x, [y] * len(x), 'k-')  # Black lines at each y-value

    #initialize values for plotting points
    y_plot_point = 103
    marker_type = 'o'

    #for each df in the list, 
    for df in df_list:
        #calculate the plot value, based on the metric name to be graphed
        if metric_name == 'Urban Indoor Coverage (Percentage)':
            plot_value = calculate_indoor_coverage(df)
        elif metric_name == 'Urban Outdoor Coverage (Percentage)':
            plot_value = calculate_outdoor_coverage(df)
        elif metric_name == 'Agricultural Coverage (Percentage)':
            plot_value = calculate_ag_coverage(df)
        elif metric_name == 'Groundwater Storage in MAF': 
            max_date_index = df['Date'].idxmax()
            plot_value = df.loc[max_date_index, 'Total_gw_storage'] / 1000000 #divide by a million to get MAF
        elif metric_name == 'Electricity Use in TWh':
            plot_value = electric_use = df['Total_GwH'].sum() / 1000 #divide by 1000 to convert GWh to TWh
        elif metric_name == 'Hydropower Generation in TWh':
            plot_value = df['Total_hydro_generation'].sum() / 1000 #divide by 1000 to convert GWh to TWh
        else:
            print('Your metric name is invalid')
            return

        #check the y plot point variable, graph appropriately, and then correct the upcoming marker values 
        if y_plot_point in [3, 23, 43, 63, 83, 103]:
            plt.plot(plot_value, y_plot_point, marker=marker_type, color='#984ea3', markersize=8)
            y_plot_point = y_plot_point -3
            marker_type = '^'
        elif y_plot_point in [0, 20, 40, 60, 80, 100]:
            plt.plot(plot_value, y_plot_point, marker=marker_type, color='#ff7f00', markersize=8)
            y_plot_point = y_plot_point -3
            marker_type = 's'
        elif y_plot_point == -3:
            #do something for the final iteration to add the refloca values to all 5 lines
            plt.plot(plot_value, y_plot_point, marker=marker_type, color='#f781bf', markersize=8)
            y_plot_point = 120
            marker_type = 'x'
        elif y_plot_point == 120:
            plt.axvline(x=plot_value, color='black', linestyle='--')
        else:
            plt.plot(plot_value, y_plot_point, marker=marker_type, color='#f781bf', markersize=8)
            y_plot_point = y_plot_point - 14
            marker_type ='o'
    
    # Customize the plot (optional)
    plt.xlabel(metric_name, fontsize=12)
    plt.xticks(fontsize=12)
    plt.yticks([0, 20, 40, 60, 80, 100], ['Managed Aquifer Recharge', 'Wastewater Recycling', 'Cropland Retirement', 'Outdoor Conservation', 'Indoor conservation', 'No Adaptation'],
              fontsize=12)  # Show y-axis ticks at specified values
    if metric_name in ['Urban Indoor Coverage (Percentage)', 'Urban Outdoor Coverage (Percentage)']:
        plt.xlim(83, 102)
    elif metric_name == 'Agricultural Coverage (Percentage)':
        plt.xlim(48, 82)
      # Set x-axis limits
    elif metric_name == 'Groundwater Storage in MAF':
        plt.xlim(1050, 1550)
    elif metric_name == 'Electricity Use in TWh':
        plt.xlim(3900, 6100) 
    else:
        plt.xlim(7400, 9100)
    plt.ylim(-10, 112)
    
    # Display the plot
    plt.show()
```

```python
def graph_monthly_change_line (df_list, plot_title, x_label, data_labels, y_label):

    # Create a figure and axes
    plt.figure(figsize=(12, 6))
    colormap= ['#984ea3', '#ff7f00', '#f781bf']

    
    for df, i in zip(df_list, range(len(df_list))):
        df = df[df['Date'].dt.year >= 2020]
        df['Month'] = df['Date'].dt.month_name().tolist()
        column_name = df.columns[2]
    
        #group, then sort the column by month
        monthly_data = df.groupby('Month')[column_name].sum()
        monthly_data = monthly_data.sort_index(key=lambda x: pd.to_datetime(x, format='%B').month)
    
        #put the data in a df
        dfmonthly_data = monthly_data.reset_index()
        plt.plot(dfmonthly_data['Month'], dfmonthly_data[column_name], color = colormap[i], label= data_labels[i])

    #put the y-axis in millions
    plt.gca().yaxis.set_major_formatter(
        plt.FuncFormatter(lambda x, pos: f'{(x/1e6)}M'))
    
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    # plt.title(plot_title)
    plt.xticks(rotation=45)  # Rotate and align x-axis labels 

    plt.legend()
    plt.show()
```
